```python
data_frame2 = pd.read_csv('/content/diabetes.csv')
data_frame2
```

```python
feature_cols = ['Pregnancies', 'Insulin', 'BloodPressure', 'Age','Glucose','BMI','DiabetesPedigreeFunction']
X = data_frame2[feature_cols] # Features
y = data_frame2.Outcome # Target variable
```

```python
feature_cols
```

```python
X
```

```python
y
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X1 = scaler.fit_transform(X)


# split X and y into training and testing sets
from sklearn.model_selection import train_test_split
X1_train,X1_test,y_train,y_test=train_test_split(X1,y,test_size=0.25,random_state=0)
```

```python
X1
```

```python
X1_train
```

```python
X1_test
```

```python
y_train
```

```python
y_test
```

```python
# import the class
from sklearn.linear_model import LogisticRegression


# instantiate the model (using the default parameters)
logreg = LogisticRegression()
logreg.fit(X1_train, y_train)


y_pred=logreg.predict(X1_test)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test,y_pred)
print (accuracy)


# import the metrics class
from sklearn import metrics
cnf_matrix = metrics.confusion_matrix(y_test, y_pred)
cnf_matrix
```

```python
# import required modules
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
class_names=[0,1] # name  of classes
fig, ax = plt.subplots()
tick_marks = np.arange(len(class_names))
plt.xticks(tick_marks, class_names)
plt.yticks(tick_marks, class_names)
# create heatmap
sns.heatmap(pd.DataFrame(cnf_matrix), annot=True, cmap="YlGnBu" ,fmt='g')
ax.xaxis.set_label_position("top")
plt.tight_layout()
plt.title('Confusion matrix', y=1.1)
plt.ylabel('Actual label')
plt.xlabel('Predicted label')
```

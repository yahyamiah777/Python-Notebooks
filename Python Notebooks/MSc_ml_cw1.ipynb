```python
import pandas as pd

df = pd.read_csv('/content/Monkeypox Coursework Dataset.csv')
```

```python
df
```

```python
columns_to_drop = ['Test ID','Home ownership','Month of Birth','Health Insurance']
```

```python
df = df.drop(columns = columns_to_drop)
```

```python
df
```

```python
print(df['MPOX PCR Result'].value_counts())
```

```python
#Category of a Nominal scale as there are only two values
df = pd.get_dummies(df, columns = ['MPOX PCR Result'], drop_first = True )
```

```python
df
```

```python
df.info()
```

```python
df.describe()
```

```python
df.hist(column = 'MPOX PCR Result_Positive')
```

```python
#find unique values
for column in df.columns:
  unique_values = df[column].unique()
  print(f"{column}: {unique_values}")
```

```python
#correcting/removing Age values
df['Age'] = df['Age'].replace({'Twenty': 20})
df['Age'] = pd.to_numeric(df['Age'], errors = 'coerce')
df = df[df['Age'] >= 0]

#removing all missing values
df = df.dropna()

#removing redundant column
del df['Systemic Illness']

#correcting Oral Lesions values
df['Oral Lesions'] = df['Oral Lesions'].replace({'YES': 1, 'No': 0})
df['Oral Lesions'] = pd.to_numeric(df['Oral Lesions'])
```

```python
for column in df.columns:
  unique_values = df[column].unique()
  print(f"{column}: {unique_values}")
```

```python
df
```

```python
df.info()
```

```python
X = df[['Encoded Systemic Illness', 'Rectal Pain','Sore Throat','Penile Oedema','Oral Lesions','Solitary Lesion','Swollen Tonsils','HIV Infection','Red blood cells count','White blood cells count','Age','Sexually Transmitted Infection']]
```

```python
X
```

```python
y = df['MPOX PCR Result_Positive']
```

```python
y
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X1 = scaler.fit_transform(X)

from sklearn.model_selection import train_test_split
X1_train,X1_test,y_train,y_test = train_test_split(X1,y,test_size = 0.25, random_state=7, stratify=y)
```

```python
from sklearn.linear_model import LogisticRegression

logreg = LogisticRegression()
logreg.fit(X1_train, y_train)

y_pred = logreg.predict(X1_test)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test,y_pred)
f1 = f1_score(y_test, y_pred)
roc_auc = roc_auc_score(y_test, logreg.predict_proba(X1_test)[:,1])

print(accuracy)
print(recall)
print(precision)
print(f1)
print(roc_auc)

from sklearn import metrics
cnf_matrix = metrics.confusion_matrix(y_test, y_pred)
cnf_matrix
```

```python
from sklearn.tree import DecisionTreeClassifier

dtc = DecisionTreeClassifier(random_state=7)
dtc = dtc.fit(X1_train,y_train)

y_pred = dtc.predict(X1_test)
```

```python
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test,y_pred)
f1 = f1_score(y_test, y_pred)
roc_auc = roc_auc_score(y_test, dtc.predict_proba(X1_test)[:,1])

print(accuracy)
#or print("Accuracy: ", metrics.accuracy_score(y_test, y_pred))
print(recall)
print(precision)
print(f1)
print(roc_auc)

cnf_matrix = metrics.confusion_matrix(y_test, y_pred)
cnf_matrix
```

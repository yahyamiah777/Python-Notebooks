# Instruction 4:

```python
#I got the minimum index, then queried it into a new dataframe,
#and then I removed that index temporarily to find the next minimum duration
#repeating the process once more
#I then concatenated the dataframes together

#TED['duration'].idxmin()
shortestTalk1 = TED.query("index == 4056")

#TED = TED.drop(4056)
#TED['duration'].idxmin()
shortestTalk2 = TED.query("index == 4793")

#TED = TED.drop(4793)
#TED['duration'].idxmin()
shortestTalk3 = TED.query("index == 4783")

shortestTalks = pd.concat([shortestTalk1,shortestTalk2, shortestTalk3])

shortestTalks
```

```python
plt.figure(figsize=(10,8), dpi=200)
sns.barplot(data=shortestTalks, x='title', y='duration', ci=None)
```
